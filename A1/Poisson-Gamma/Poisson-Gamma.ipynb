```python
import numpy as np
import scipy
import scipy.stats
import matplotlib as mpl   
import matplotlib.pyplot as plt
import statistics
%matplotlib inline
```

```python
transition_model = lambda x: np.random.normal(x, [0.05, 5], (2,))
import math
def prior(w):
    if w[0] <= 0 or w[1] <= 0:
        return 0
    else:
        return 1
    
def manual_log_lik_gamma(x, data):
    return np.sum((x[0] - 1) * np.log(data) - (1 / x[1]) * data - x[0] * np.log(x[1]) - np.log(math.gamma(x[0])))
    
def log_lik_gamma(x, data):
    return np.sum(np.log(scipy.stats.gamma(a = x[0], scale = x[1], loc = 0).pdf(data)))

def log_lik_poisson(x, data):
    return np.sum(scipy.stats.poisson.logpmf(data, x))

def acceptance(x, x_new):
    if x_new > x:
        return True
    else:
        accept = np.random.uniform(0, 1)
        # Since we did a log likelihood, we need to exponentiate in order to compare to the random number
        # less likely x_new are less likely to be accepted
        return (accept < (np.exp(x_new-x)))
```

```python
def metropolis_hastings(likelihood_computer,
                        prior,
                        transition_model,
                        param_init,
                        iterations,
                        data,
                        acceptance_rule):
    # likelihood_computer(x,data): returns the likelihood that these parameters generated the data
    # transition_model(x): a function that draws a sample from a symmetric distribution and returns it
    # param_init: a starting sample
    # iterations: number of accepted to generated
    # data: the data that we wish to model
    # acceptance_rule(x,x_new): decides whether to accept or reject the new sample
    if likelihood_computer == log_lik_poisson:
        data_lik = [round(i) for i in data]
    x = param_init
    accepted = [param_init]
    rejected = []
    alltests = [param_init]
    for i in range(iterations):
        x_new = transition_model(x)
        if likelihood_computer == log_lik_poisson:
            x_lik = likelihood_computer(x[0]/x[1], data_lik)
            x_new_lik = likelihood_computer(x_new[0]/x_new[1], data_lik)
        else:
            x_lik = likelihood_computer(x, data)
            x_new_lik = likelihood_computer(x_new, data)
        alltests.append(x_new)
        if (acceptance_rule(x_lik + np.log(prior(x)), x_new_lik + np.log(prior(x_new)))):            
            x = x_new
            accepted.append(x_new)
        else:
            rejected.append(x_new)
        
        if i % 10000 == 9999:
            print(i + 1)
                
    return np.array(alltests), np.array(accepted), np.array(rejected)
```

```python
competitions = ['', ' - SA', ' - SB', ' - SC', ' - SD', ' - CdB']

for competition in competitions:
    goals = np.loadtxt("..\\all goals{}.csv".format(competition), delimiter=',')
    x_init = [1, 1]
    iterations = 100000
    inc = 0.1
    figure = 1
    first = 500
    last = 500
    years = goals[:, 1]
    activity = goals[:, 2] + inc # inc para evitar log de 0

    fig = plt.figure(figsize = (10, 5))
    ax = fig.add_subplot(1, 1, 1)
    ax.plot(years, activity, label = "Gols", color = "#1F77B4")
    ax.set_yticks(np.arange(0, 11, 1))
    ax.set_xlabel("Jogo")
    ax.set_ylabel("Gols")
    # ax.set_title("Figura {}: Gols por jogo no perodo de 2013 a 2020 na Srie A do Campeonato Brasileiro".format(figure))
    figure += 1
    plt.savefig('fig01{} - poissongamma.png'.format(competition))
    print()


    fig = plt.figure(figsize = (10, 10))
    ax = fig.add_subplot(1, 1, 1)
    ax.hist(activity, bins = np.linspace(0, 10, 11), density = True, color = "#1F77B4")
    ax.set_xticks(np.arange(0, 10, 1))
    ax.set_xlabel("Contagem de gols")
    ax.set_ylabel("Frequncia")
    # ax.set_title("Figura {}: Histograma mostrando a frequncia de gols por jogo nos 8 anos (2013-2020)".format(figure))
    figure += 1
    plt.savefig('fig02{} - poissongamma.png'.format(competition))
    print()


    model = lambda t, x : np.random.poisson(x[0]/x[1], t)
    t = np.arange(activity.shape[0])
    observation_gen = model(t.shape[0], x_init)
    fig = plt.figure(figsize = (10, 10))
    ax = fig.add_subplot(1, 1, 1)

    ax.hist(observation_gen, bins = np.linspace(0, 10, 11), density = True, label = "Valores preditos", color = "#FFBE86")
    ax.hist(activity, bins = np.linspace(0, 10, 11), alpha = 0.5, density = True, label = "Valores originais", color = "#1F77B4")
    ax.set_xlabel("Gols")
    ax.set_ylabel("Frequncia")
    ax.set_xticks(np.arange(0, 11, 1))
    # ax.set_title("Figura {}: Distribuio a priori das predies".format(figure))
    figure += 1
    plt.savefig('fig03{} - poissongamma.png'.format(competition))
    ax.legend()
    print()


    alltests, accepted, rejected = metropolis_hastings(log_lik_poisson,
                                                       prior,
                                                       transition_model,
                                                       x_init,
                                                       iterations,
                                                       activity,
                                                       acceptance)


    # PELA MDIA DE GOLS
    data_lik = [round(i) for i in activity]
    mu = np.array(data_lik).mean()
    sigma2 = statistics.pvariance(data_lik)
    alpha = mu**2/sigma2
    beta = mu/sigma2
    model = lambda t, alpha, beta: np.random.poisson(alpha/beta, t)
    t = np.arange(activity.shape[0])
    observation_gen = model(t.shape[0], alpha, beta)
    fig = plt.figure(figsize = (10, 10))
    ax = fig.add_subplot(1, 1, 1)

    ax.hist(observation_gen, bins = np.linspace(0, 10, 11), density = True, label = "Valores preditos", color = "#FFBE86")
    ax.hist(activity, bins = np.linspace(0, 10, 11), alpha = 0.5, density = True, label = "Valores originais", color = "#1F77B4")
    ax.set_xlabel("Gols")
    ax.set_ylabel("Frequncia")
    ax.set_xticks(np.arange(0, 11, 1))
    # ax.set_title("Figura {}: Distribuio a posteriori das predies".format(figure))
    figure += 1
    plt.savefig('fig04{} - poissongamma.png'.format(competition))
    ax.legend()
    print()


    a = []
    r = []
    for i in range(len(alltests)):
        if alltests[i] in accepted:
            a.append(alltests[i])
        else:
            r.append(alltests[i])


    aceitos = np.array(a)
    rejeitados = np.array(r)

    fig = plt.figure(figsize = (10, 10))
    ax = fig.add_subplot(2, 1, 1)
    ax.plot(aceitos[:, 0], aceitos[:, 1], 'b.', label = 'Aceitos', color = "#1F77B4")
    ax.plot(rejeitados[:, 0], rejeitados[:, 1], 'rx', label = 'Rejeitados')
    ax.set_xlabel("$a$")
    ax.set_ylabel("$b$")
    ax.legend()

    ax = fig.add_subplot(2, 1, 2)
    ax.plot(aceitos[:, 0], aceitos[:, 1], label = "Caminho", color = "#1F77B4")
    ax.plot(aceitos[:, 0], aceitos[:, 1], 'b.', label = 'Aceitos', color = "#1F77B4")
    ax.set_xlabel("$a$")
    ax.set_ylabel("$b$")
    plt.savefig('fig05{} - poissongamma.png'.format(competition))
    print()


    show = int(0.8 * len(alltests))
    a = []
    r = []
    for i in range(len(alltests) - show + 1, len(alltests)):
        if alltests[i] in accepted:
            a.append(alltests[i])
        else:
            r.append(alltests[i])


    aceitos = np.array(a)
    rejeitados = np.array(r)

    fig = plt.figure(figsize = (10, 10))
    ax = fig.add_subplot(2, 1, 1)
    ax.plot(aceitos[:, 0], aceitos[:, 1], 'b.', label = 'Aceitos', color = "#1F77B4")
    ax.plot(rejeitados[:, 0], rejeitados[:, 1], 'rx', label = 'Rejeitados')
    ax.set_xlabel("$a$")
    ax.set_ylabel("$b$")
    ax.legend()

    ax = fig.add_subplot(2, 1, 2)
    ax.plot(aceitos[:, 0], aceitos[:, 1], label = "Caminho", color = "#1F77B4")
    ax.plot(aceitos[:, 0], aceitos[:, 1], 'b.', label = 'Aceitos', color = "#1F77B4")
    ax.set_xlabel("$a$")
    ax.set_ylabel("$b$")
    plt.savefig('fig06{} - poissongamma.png'.format(competition))
    print()


    fig = plt.figure(figsize = (10, 5))
    ax = fig.add_subplot(1, 2, 1)
    ax.plot(aceitos[:, 0], 'b.', label = 'Aceitos', color = "#1F77B4")
    # ax.set_title("Figura {}: Trao de $a$".format(figure))
    figure += 1
    ax.set_xlabel("Iterao")
    ax.set_ylabel("$a$")
    ax = fig.add_subplot(1, 2, 2)
    data, binsa, alpha = ax.hist(aceitos[:, 0], bins = 20, density = True, color = "#1F77B4")
    ax.set_ylabel("Frequncia")
    ax.set_xlabel("$a$")
    # ax.set_title("Figura {}: Histograma de $a$".format(figure))
    figure += 1
    fig.tight_layout()
    plt.savefig('fig07{} - poissongamma.png'.format(competition))
    
    fig = plt.figure(figsize = (10, 5))
    ax = fig.add_subplot(1, 2, 1)
    ax.plot(aceitos[:, 1], 'b.', label = 'Aceitos', color = "#1F77B4")
    # ax.set_title("Figura {}: Trao de $a$".format(figure))
    figure += 1
    ax.set_xlabel("Iterao")
    ax.set_ylabel("$b$")
    ax = fig.add_subplot(1, 2, 2)
    data, binsb, alpha = ax.hist(aceitos[:, 1], bins = 20, density = True, color = "#1F77B4")
    ax.set_ylabel("Frequncia")
    ax.set_xlabel("$b$")
    # ax.set_title("Figura {}: Histograma de $a$".format(figure))
    figure += 1
    fig.tight_layout()
    plt.savefig('fig08{} - poissongamma.png'.format(competition))
    
    fig = plt.figure(figsize = (10, 7))
    ax = fig.add_subplot(1, 1, 1)
    xbins = np.linspace(min(binsa) - .01, max(binsa) + .01, 75)
    ybins = np.linspace(min(binsb) - .1, max(binsb) + .1, 75)
    counts, xedges, yedges, im = ax.hist2d(aceitos[:,0], aceitos[:,1], density = True, bins = [xbins, ybins])
    ax.set_xlabel("a")
    ax.set_ylabel("b")
    fig.colorbar(im, ax = ax)
    # ax.set_title("2D histogram showing the joint distribution of $a$ and $b$")
    plt.savefig('fig09{} - poissongamma.png'.format(competition))


    # PELA MDIA DAS LTIMAS ITERAES
    alpha = aceitos[:, 0].mean()
    beta = aceitos[:, 1].mean()
    model = lambda t, alpha, beta : np.random.poisson(alpha/beta, t)
    t = np.arange(activity.shape[0])
    observation_gen = model(t.shape[0], alpha, beta)
    fig = plt.figure(figsize = (10, 10))
    ax = fig.add_subplot(1, 1, 1)

    ax.hist(observation_gen, bins = np.linspace(0, 10, 11), density = True, label = "Valores preditos", color = "#FFBE86")
    ax.hist(activity, bins = np.linspace(0, 10, 11), alpha = 0.5, density = True, label = "Valores originais", color = "#1F77B4")
    ax.set_xlabel("Gols")
    ax.set_ylabel("Frequncia")
    ax.set_xticks(np.arange(0, 11, 1))
    # ax.set_title("Figura {}: Distribuio a posteriori das predies".format(figure))
    figure += 1
    plt.savefig('fig10{} - poissongamma.png'.format(competition))
    ax.legend()
    print()
```
